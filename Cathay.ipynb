```python

page = chrome_browser.page_source
page_soup = BeautifulSoup(page,'html.parser')
print(page_soup)
```

```python
containers = page_soup.findAll("h3", {"class" : "title-article-light"})
# print(containers)
print(len(containers))
print(type(containers))
```

```python
print(containers[1].text)
```

```python
prod = []
for i in range(1,len(containers)):
    prod.append(containers[i].text)
prod
```

```python
print(prod[3][-4:-1])
firstindex = prod[3].find(":")
lastindex = prod[3].find(")")
print(firstindex)
print(lastindex)
print(prod[3][firstindex+1:lastindex])
print(prod[3][:firstindex-5])
```

```python
prod_name = []
prod_code = []
for i in range(len(prod)):
    firstindex = prod[i].find(":")
    lastindex = prod[i].find(")")
    prod_name.append(prod[i][:firstindex-5])
    prod_code.append(prod[i][firstindex+1:lastindex])
prod_name
prod_code
```

```python
df = pd.DataFrame({"商品名稱":prod_name,"商品代碼":prod_code})
df
```
