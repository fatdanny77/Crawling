```python
import requests
from bs4 import BeautifulSoup

url = "http://exam.lib.ntu.edu.tw/graduate/term/304%201330%20191%20189%20190"
re = requests.get(url)
soup = BeautifulSoup(re.text, "html.parser")
table = soup.table.tbody

for each_row in table.findAll("tr"):
    td = each_row.findAll("td")
    year = td[0].text.strip()
    dep = td[1].text.strip()
    course = td[2].text.strip()
    link = td[3].a["href"].split("?file=")[1].split("&")[0]
    print(year, dep, course)
    print(link)
    #file_name = year + "_" + dep + "_" + course
    #print(file_name)
    
    file_name = f"{year}_{dep}_{course}.pdf"
    print(file_name)
    f = open(file_name, "wb")
    r = requests.get(link)
    f.write(r.content)
    r.close()
    f.close()
```
